```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1) 

df_nona = df.dropna()

df_nona
 
#2)

from sklearn.model_selection import train_test_split

y = df_nona['Label']
X_train, X_test, y_train, y_test = train_test_split(df_nona, y, test_size=0.30, random_state = 0)

# Check if split is actually correct
# We can see it's roughly 70% train and 30% test
# So we can proceed!
print(df_nona.shape[0])
print(float(X_train.shape[0]) / float(df_nona.shape[0]))
print(float(X_test.shape[0]) / float(df_nona.shape[0]))

#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```
