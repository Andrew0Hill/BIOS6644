```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1)
df = df.dropna(how="any")
df.head()
```

```python
 
#2)

from sklearn.model_selection import train_test_split

y = df['Label']
df_features = df.iloc[:,1:]

X_train, X_test, y_train, y_test = train_test_split(df_features, y, test_size=0.3)

#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
```

```python
predictions
```

```python
reality
```

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(reality, predictions)
print('Confusion Matrix : \n', cm)

total=sum(sum(cm))
# calculate accuracy
accuracy=(cm[0,0]+cm[1,1])/total
print ('Accuracy : ', accuracy)

# calculate sensitivity: correctly identifying true positives (controls)
sensitivity = cm[0,0]/(cm[0,0]+cm[0,1])
print('Sensitivity : ', sensitivity )

# calculate specificity: correctly identifying true negatives (Chronic Kidney Disease)
specificity = cm[1,1]/(cm[1,0]+cm[1,1])
print('Specificity : ', specificity)

# got code from: https://statinfer.com/204-4-2-calculating-sensitivity-and-specificity-in-python/

#######################################   END STUDENT CODE   #####################################################
```
