## Read Data with Pandas
If you're going to do some exploratory data analysis, Pandas has wonderful tools including a robust data reader.

In this exercsise, we'll use the Educational Attainment data file mentioned on Canvas [above](https://data-cdphe.opendata.arcgis.com/datasets/cee9e5ea911c4814999de34806474105_3/data):

 

```python
import pandas as pd
import pylab as plt
%matplotlib notebook
```

```python
# Read the data
datafile = 'Data/Educational_Attainment_Census_Tracts.csv'
df = pd.read_csv(datafile)
```

```python
# Now our data are in a pandas data frame.  You can get a "pretty" view just by printing df
df

# Notice that the column Tract_Name automatically handled the quoted field with commas inside.
```

```python
# We can get a county list in a similarly pithy bit of code
# that works on the data frame

counties = [xx.capitalize() for xx in list(set(df.County))]
counties.sort()
print(", ".join(counties))
```

```python
# Data Frames have a describe() method that gives a nice summary:
df.describe()
```

```python
# Nominal data fields can be nicely 
# summarized with value_counts()
df.County.value_counts()
```

```python
plt.plot(df.Population_Density_PerLandSquareMile, df.Percent_Education_Population_AgeOver24_NoHSDiploma,'.')
```

```python
plt.xlabel('Density')
plt.ylabel('% >= BS Degree')
```
